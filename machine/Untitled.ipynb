```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-5,5,0.1)
y1=x**2 -2
y2=20*np.cos(x)**2

plt.figure(1) #1번 그래프 창을 생성
plt.plot(x,y1)# 지정된 그래프 창에 그래프 그림

plt.figure(2) #2번 그래프 창을 생성
plt.plot(x,y2)#지정된 그래프 창에 그래프 그림

plt.figure(1) #이미 생성된 1번 그래프 창을 지정
plt.plot(x,y2)# 지정된 그래프 창에 그래프 그림

plt.figure(2)#이미 생성된 2번 그래프 창을 지정
plt.clf()    # 지정된 그래프 창에 그려진 그래프를 모두 지움
plt.plot(x,y1,'r,-.')# 지정된 그래프 창에 그래프 그림
plt.show()
```

```python
import seaborn as sns
font_location= '/Users/chan/Library/Fonts/BMEULJIRO.otf'
font_name = font_manager.FontProperties(fname=font_location).get_name()

#iris = sns.load_dataset("iris") #붓꽃 데이터
#iris = sns.load_dataset("iris") #붓꽃 데이터
#iris = sns.load_dataset("iris") #붓꽃 데이터
iris = sns.load_dataset("iris") #붓꽃 데이터
x = iris.petal_length.values
plt.rcParams["font.family"] = font_name
sns.kdeplot(x)
plt.title("Iris 데이터 중 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```
