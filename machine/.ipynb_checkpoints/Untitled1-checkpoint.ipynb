```python
from sklearn.preprocessing import normalize
import seaborn as sns
```

```python
x= [[1.,-1.,2.],
    [2.,0.,0.,],
    [0.,1.,-1.,]]
x_normalized_l1 = normalize(x, norm='l1') # 1차식
x_normalized_l1
```

```python

x_normalized_l2 = normalize(x, norm='l2') # 2차식
x_normalized_l2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
iris = sns.load_dataset("iris")
y = iris.species
le = LabelEncoder()
le.fit(y)
species = le.transform(y)
species
```

```python
iris
```

```python
y
```

```python
les = le.inverse_transform(species)
les
```

```python
species
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc.fit(species.reshape(-1,1))
```

```python
species.reshape(-1,1)
```

```python
iris_onehot = enc.transform(species.reshape(-1,1))
iris_onehot
iris_onehot.toarray()
```
